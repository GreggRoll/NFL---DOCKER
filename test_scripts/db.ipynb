```python
import sqlite3

def create_data_log_database():
    conn = sqlite3.connect('../data-log.db')  # This will create the database file
    cursor = conn.cursor()

    # Create a table for NFL data
    cursor.execute(''' 
    CREATE TABLE IF NOT EXISTS bovada_data (
        date DATE,
        time TEXT,
        bets INTEGER,
        home_team TEXT,
        away_team TEXT,
        home_win INTEGER,
        away_win INTEGER,
        win_differential INTEGER,
        day TEXT,
        points REAL,
        game_id TEXT
    )
    ''')

    conn.commit()
    conn.close()
    

def drop_bovada():
    conn = sqlite3.connect('../data-log.db')  # This will create the database file
    cursor = conn.cursor()

    # Create a table for NFL data
    cursor.execute('DROP TABLE bovada_data')
    conn.commit()
    conn.close()

def create_matchup_database():
    conn = sqlite3.connect('../data-log.db')  # This will create the database file
    cursor = conn.cursor()

    # Create a table for NFL data
    cursor.execute(''' 
    CREATE TABLE IF NOT EXISTS matchup_data (
        time Text,
        matchup TEXT,
        projected_winner TEXT,
        ranking INTEGER,
        game_id TEXT
    )
    ''')

    conn.commit()
    conn.close()


def drop_table(table):
    conn = sqlite3.connect('../data-log.db')  # This will create the database file
    cursor = conn.cursor()

    # Create a table for NFL data
    cursor.execute(f''' 
    DROP TABLE {table}
    ''')

    conn.commit()

    conn.close()

def create_merge_database():
    conn = sqlite3.connect('../data-log.db')  # This will create the database file
    cursor = conn.cursor()

    # Create a table for NFL data
    cursor.execute(''' 
    CREATE TABLE IF NOT EXISTS merged_data (
        game_id TEXT,
        matchup TEXT,
        projected_winner TEXT,
        ranking INTEGER,
        alt_game_id TEXT,
        week TEXT,
        Game TEXT,
        Time TEXT,
        pct TEXT,
        message TEXT,
        IngestTime TEXT
    )
    
    ''')

    conn.commit()
    conn.close()


# drop_bovada()

drop_table('espn_news')
# create_data_log_database()
# create_matchup_database()
# create_merge_database()
```

```python
def create_news():
    # Connect to SQLite database
    conn = sqlite3.connect('../data-log.db')
    cursor = conn.cursor()

    # Create table if it doesn't exist
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS espn_news (
        title TEXT PRIMARY KEY,
        date DATE,
        link TEXT,
        image_url TEXT,
        relevant TEXT,
        ai_score TEXT
    )
    ''')

create_news()
```
